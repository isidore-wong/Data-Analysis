对在线零售数据进行RFM分析及可视化呈现

```python
import pandas as pd
import numpy as np
import datetime
import math
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文标签
# plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
# 读取zip压缩包中的数据文件
def get_zip(file_name):
    with zipfile.ZipFile(file_name, 'r') as z:
        f = z.open('data.csv')
        data = pd.read_csv(f, index_col=0)
    return data
```

```python
file_name = '../../00_dataset/OnlineRetailDataset.zip'
df = get_zip(file_name)
df_data = df.copy()
df_data.head()
```

#### 数据字段的说明
InvoiceNo: 发票号码，每笔交易分配唯一的6位整数，而退货订单的代码以字母'c'开头

StockCode: 产品代码，分配唯一代码

Description: 产品描述，对每件产品的简略描述

Quantity: 产品数量，每笔交易中每件产品的数量

InvoiceDate: 发票日期和时间，每笔交易发生的日期和时间

UnitPrice: 单价（英镑），单位产品价格

CustomerID:顾客号码，每个客户分配唯一的5位整数

Country: 国家的名字，每个客户所在国家/地区的名称

```python
df_data.info()
```

```python
# 查看缺失CustomerID的数据
df_customer_null = df_data.loc[df_data['CustomerID'].isnull(), :]
df_customer_null.head()
df_customer_null.count()
```

```python
# InvoiceDate字段进行转化为日期类型
df_data['InvoiceDate'] = pd.to_datetime(df_data['InvoiceDate'])
data_info = pd.DataFrame(df_data.dtypes).T.rename({0: 'column type'})
data_info = data_info.append(pd.DataFrame(df_data.isnull().sum()).T.rename(index={0: 'null values(num)'}))
data_info = data_info.append(pd.DataFrame(df_data.isnull().sum() / df_data.shape[0] * 100).T.rename(index={0: 'null values(%)'}))
display(data_info)
# display(df_data[:5])
```

```python
# 删除重复值
df_data = df_data.drop_duplicates()
df_data.describe()    # Quantily/UnitPrice这两列最小值为负数，需要去除这些异常值
df_data = df_data[df_data['Quantity'] > 0]
df_data = df_data[df_data['UnitPrice'] > 0]
df_data.describe()
```

```python
# 删除没用用户ID的数据
df_data = df_data.dropna(subset=['CustomerID'])

# 将清洗后的数据输出
df_data.to_csv('retail_data_cleaned.csv')
```

```python
# R值的构造
refrence_date = df_data['InvoiceDate'].max() + datetime.timedelta(days=1)  # 设置参考日期为样本中最大日期+1
df_r = df_data.groupby('CustomerID')['InvoiceDate'].max().reset_index()
df_r.columns = ['CustomerID', 'LastPurchasedate']
df_r['DaysSinceLastPurchase'] = (refrence_date - df_r['LastPurchasedate']).astype('timedelta64[D]')
# df_r['DaysSinceLastPurchase'].describe()

# F值的构造
dup_f= df_data.groupby(['InvoiceNo', 'CustomerID'])['InvoiceDate'].count().reset_index()
df_freq = dup_f.groupby('CustomerID')['InvoiceDate'].count().reset_index()
df_freq.columns = ['CustomerID', 'Frequency']
df_rf = df_r.merge(df_freq)
df_rf.head()

# M值的构造
df_data['TotalMoney'] = df_data['Quantity'] * df_data['UnitPrice']
df_m = df_data.groupby('CustomerID')['TotalMoney'].sum().reset_index()
df_rfm = df_rf.merge(df_m)
df_rfm = df_rfm[['CustomerID', 'DaysSinceLastPurchase', 'Frequency', 'TotalMoney']]
df_rfm.columns = ['CustomerID', 'R', 'F', 'M']
df_rfm.head()
```

```python
# df_data[df_data['CustomerID'] == 12346]

# 对构造的RFM进行打分，采用四分位进行打分
quantiles = df_rfm.quantile(q=[0.25, 0.5, 0.75])
quantiles = quantiles.to_dict()
segmented_rfm = df_rfm.copy()
segmented_rfm.head()

# 为RFM打分
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1

def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
    

segmented_rfm['R-Score'] = segmented_rfm['R'].apply(RScore, args=('R', quantiles, ))
segmented_rfm['F-Score'] = segmented_rfm['F'].apply(FMScore, args=('F', quantiles, ))
segmented_rfm['M-Score'] = segmented_rfm['M'].apply(FMScore, args=('M', quantiles, ))

# segmented_rfm.head()

segmented_rfm['RFM-Score'] = segmented_rfm['R-Score'] * 100 + segmented_rfm['F-Score'] * 10 + segmented_rfm['M-Score']

# 筛选出RFM分数为'444'的用户
# segmented_rfm[segmented_rfm['RFM-Score'] == 444].sort_values('M', ascending=False).head(10)
```

```python
# 对客群进行分组
score_level = pd.read_excel('Score_Level.xlsx')
score_level
```

#### RFM分组依据
- 444-433：最有价值用户，活跃度高、消费频次和消费贡献高，可主推高客单价/新品
- 433-332：处于沉默阶段、但消费能力较强的用户，可对于爆品等产品进行再营销
- 332-222：最近未购买的价值客户，需要进一步促销激活，进行价格刺激
- 222-111：价值最低的用户，主要是价格敏感型用户

```python
segmented_rfm['RFM-Score'] = segmented_rfm['RFM-Score'].astype(int)

# 对RFM-Score进行分组
def rfm_level(RFMScore):
    if (RFMScore >= 433 and RFMScore < 444):
        return 4
    elif (RFMScore >= 322 and RFMScore < 433):
        return 3
    elif (RFMScore >= 222 and RFMScore < 332):
        return 2
    return 1

segmented_rfm['RFM-Level'] = segmented_rfm['RFM-Score'].apply(rfm_level)
segmented_rfm.head()
```

```python
plt.figure(figsize=(10, 4))
color = sns.color_palette()
sns.countplot(x='RFM-Level',data=segmented_rfm, color = color[9])
# plt.bar(segmented_rfm['RFM-Score'], segmented_rfm['RFM-Score'].count(), width=0.2, color='g')
plt.xlabel('score level')
plt.ylabel('score count')
plt.title('RFM level distribution')
plt.show()
```

```python
from sklearn import preprocessing
# 采用K-Means方法分类RFM
segmented_rfm.head()
segmented_rfm.describe()

# 对rfm-score进行归一化处理
segmented_rfm['R-log'] = segmented_rfm['R'].apply(math.log)
segmented_rfm['F-log'] = segmented_rfm['F'].apply(math.log)
segmented_rfm['M-log'] = segmented_rfm['M'].apply(math.log)
feature_vector = segmented_rfm[['R-log', 'F-log', 'M-log']]
scaler = preprocessing.StandardScaler().fit(feature_vector)
rfm_scaler = scaler.transform(feature_vector)
# rfm_scaler = pd.DataFrame(rfm_scaler)
rfm_scaler
```

```python
from sklearn.cluster import KMeans

# 确定初始的质心数、达到收敛的条件、样本点到质心的平方距离和等
cl = 50    # 质心数
corte = 0.1
anterior = 1000000000000
cost = []
k_best = cl
file_txt = open('cluster_model_data.txt', 'w')

for k in range(1, cl+1):
    
    # 使用k集群创建kmeans模型
    model = KMeans(
        n_clusters=k,
        init='k-means++',    # random
        n_init=10,
        max_iter=300,
        tol=1e-04,
        random_state=101
    )
    
    model = model.fit(feature_vector)
    
    # 聚类标签
    labels = model.labels_
    # 样本点到质心的平方距离和
    interia = model.inertia_
    # print('k值及误差为：', k, ':', interia)
    # print('聚类标签: ', labels)
    file_txt.write('k值为：')
    file_txt.write(str(k))
    file_txt.write('：')
    file_txt.write('\n 误差为：')
    file_txt.write(str(interia))
    file_txt.write('\n 聚类标签: ')
    file_txt.write(str(labels))
    
    
    # 确定k的最佳值
    if(k_best == cl) and (((anterior-interia)/anterior) < corte):
        k_best = k - 1
    cost.append(interia)
    anterior = interia
    # print('簇中心：', model.cluster_centers_)
    # print(k_best)
    file_txt.write('\n 簇中心：')
    file_txt.write(str(model.cluster_centers_))
    file_txt.write('\n')
    file_txt.write(str(k_best))
    file_txt.write('\n **********-------------------------------********** \n')
    

plt.figure(figsize=(10, 6))
plt.scatter(range(1, cl+1), cost, c='red')
plt.show()
```

```python
# 通过kmeans模型得到的最佳的k值
print('最佳的k值建议为：', k_best)

model = KMeans(n_clusters=k_best, init='k-means++', n_init=10, 
               max_iter=300, tol=1e-04, random_state=101)
# 使用缩放使数据进行缩放后，能达到较好的聚类效果
model = model.fit(feature_vector)
labels_clu = model.labels_

# 聚类效果可视化
# plt.scatter(rfm_scaler[:, 0], rfm_scaler[:, 1], c=model.labels_.astype(float))
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(121)
plt.scatter(x=rfm_scaler[:, 0], y=rfm_scaler[:, 2], c=model.labels_.astype(float))
ax.set_xlabel('R-log')
ax.set_ylabel('M-log')
ax = fig.add_subplot(122)
plt.scatter(x = rfm_scaler[:, 1], y = rfm_scaler[:, 2], c=model.labels_.astype(float))
ax.set_xlabel('F-log')
ax.set_ylabel('M-log')

plt.show()
```

```python
# 3D plot
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure(figsize=(20, 12))
ax1 = fig.gca(projection='3d')

xs = rfm_scaler[:, 0]
ys = rfm_scaler[:, 1]
zs = rfm_scaler[:, 2]
ax1.scatter(xs, ys, zs, c=model.labels_.astype(float))
ax1.set_xlabel('R-log')
ax1.set_ylabel('F-log')
ax1.set_zlabel('M-log')

plt.show()
```
