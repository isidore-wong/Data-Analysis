```python
missing_data = airline_data.isnull().sum()
missing_data = missing_data[missing_data > 0]
missing_data.sort_values(inplace=True)
print(missing_data)
missing_data.plot.bar()
```

```python
# 通过观测可知，数据集中存在票价为零但是飞行公里大于零的不合理值
# 只保留票价非零的，或者平均折扣率与总飞行公里数同时为0的数据
air_data = airline_data[airline_data['SUM_YR_1'].notnull() & airline_data['SUM_YR_2'].notnull()]
clean_condition1 = airline_data['SUM_YR_1'] != 0
clean_condition2 = airline_data['SUM_YR_2'] != 0
# 该规则是删去脏数据，要么有折扣有公里数，要么就都是折扣为0且里程数为0
clean_condition3 = (airline_data['SEG_KM_SUM'] == 0) & (airline_data['avg_discount'] == 0)
air_data = air_data[clean_condition1 | clean_condition2 | clean_condition3]
print('清洗前数据记录数：{} \n清洗后数据记录数：{} \n缺失率为：{:.2f}%'.format(len(airline_data), len(air_data), 
                                                         (len(airline_data)-len(air_data))/len(airline_data)*100))
print(air_data.shape)
```

### 数据特征转换
在客户分类中，RFM模型是一个经典的分类模型，——最近消费(Recency)、消费频率(Frequency)、消费金额(Monetary)细分客户群体，从而分析不同群体的客户价值。

通过L(客户入会时间距今月数)、R(客户最近一次乘机时间距今月数)、F(客户在样本时间内乘机次数)、M(客户在样本时间内的飞行里程)、C(客户在样本时间内乘坐舱位的平均折扣系数)
因此，我们选择入会时间（FFP_DATE）、样本观测结束时间(LOAD_TIME)、飞行次数（FLIGHT_COUNT）、平均折扣率（AVG_DISCOUNT）、总飞行公里数（SEG_KM_SUM）、最后一次乘机时间距今时长（LAST_TO_END）这6个属性进行分析，删去其他弱相关属性，并将相关数据进行转换

```python
'''
L=LOAD_TIME-FFP_DATE(入会时长)
R=LAST_TO_END（最近一次乘机时间）
F=FLIGHT_COUNT（频次）
M=SEG_KM_SUM(飞行公里数)
C=AVG_DISCOUNT（平均折扣率）
'''
# 将LOAD_TIM、FFP_DATE属性的object类型转换为datetime类型
air_data['LOAD_TIME'] = pd.to_datetime(air_data['LOAD_TIME'])
air_data['FFP_DATE'] = pd.to_datetime(air_data['FFP_DATE'])
air_data['L'] = (air_data['LOAD_TIME'] - air_data['FFP_DATE']).dt.days
customer_features = air_data[['L', 'LAST_TO_END', 'FLIGHT_COUNT', 'SEG_KM_SUM', 'avg_discount']]
customer_features = customer_features.rename(
                    columns = {'L':'L入会时间(天)','LAST_TO_END':'R最近消费距今时间(天)','SEG_KM_SUM':'M总里程','FLIGHT_COUNT':'F飞行次数','avg_discount':'C平均折扣率'},
                    inplace = False
)

customer_features.head()
print('用户价值特征分析：\n', customer_features.describe())
```

```python
# 对customer_features中的数据进行标准化处理，以Z值表示
customer_Z = (customer_features - customer_features.mean(axis=0)) / (customer_features.std(axis=0))
customer_Z.columns = ['Z' + i for i in customer_features.columns]
customer_Z.head()
```

```python
'''
采用K-means进行模型构建，首先根据上述5个客户指标进行聚类，其次根据结果结合业务针对不同客户群进行分析，并进行排名
1. K-Means算法主要就在于k=n_cluster参数的确定上面，这里采用计算SSE的方法，尝试找到最好的K数值
2. SSE值代表了每一个数据点离聚类中心的距离的评分，即为误差平方。当k小于真实聚类数时，随着k增大数据之间的聚合度会增加，
SSE下降幅度会很大，当k快达到真实聚类值时，SSE下降幅度会变平稳
'''
def distEclud(vecA, vecB):
    '''
    计算两个向量的欧式距离的平方，并返回
    '''
    return np.sum(np.power(vecA - vecB, 2))

def get_nclusters(data):
    '''
    计算不同的K值，SSE的大小变化，从而确定合适的K值
    param data: 待聚类的数据
    '''
    data_cluster = data.values
    nums = range(2, 15)
    SSE = []
    for num in nums:
        sse = 0
        kmodel = KMeans(n_clusters=num, n_jobs=4)
        kmodel.fit(data_cluster)
        # 簇中心
        cluster_center_list = kmodel.cluster_centers_
        # print('簇中心为：\t', cluster_center_list)
        # 每个样本属于的簇序号列表
        cluster_list = kmodel.labels_.tolist()
        # print('簇序号列表为：\t', cluster_list)
        
        for index in range(len(data_cluster)):
            cluster_num = cluster_list[index]
            sse += distEclud(data_cluster[index, :], cluster_center_list[cluster_num])
        print('簇数为', num, '时', 'SSE是:', sse)
        SSE.append(sse)
    return nums, SSE

nums, SSE = get_nclusters(customer_Z)
```

```python
# 画图，通过观察SSE与K值的关系找到最合适的K值
# 使用ggplot的绘图风格
plt.style.use('ggplot')
# 绘图观测SSE与簇个数的关系
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(nums, SSE, marker='+')
ax.set_xlabel("n_clusters", fontsize=10)
ax.set_ylabel("SSE", fontsize=10)
fig.suptitle("KMeans", fontsize=14)
plt.show()
```

```python
# 观察图像，并没有的所谓的“肘”点出现，基本上是随k值的增大逐渐减小的。在k=8后变化放缓
# 设K=8
kmodel = KMeans(n_clusters=8, n_jobs=4)
kmodel.fit(customer_Z)
# 统计各个类别的数目，找出聚类中心
r1 = pd.Series(kmodel.labels_).value_counts()
r2 = pd.DataFrame(kmodel.cluster_centers_)
# 所有簇中心坐标值中最大值和最小值
max_value = r2.values.max()
min_value = r2.values.min()
r = pd.concat([r2, r1], axis=1)  # 横向连接（0是纵向），得到聚类中心对应的类别下的数目
r.columns = list(customer_Z.columns) + ['类别数目']  # 重命名表头
```

```python
# 绘图
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, polar=True)
center_num = r.values
feature = ["L入会时间", "R最近消费距今时间", "M总里程", "F飞行次数程", "C平均折扣率"]
N =len(feature)
for i, v in enumerate(center_num):
    # 设置雷达图的角度，用于平分切开一个圆面
    angles=np.linspace(0, 2*np.pi, N, endpoint=False)
    # 为了使雷达图一圈封闭起来，需要下面的步骤
    center = np.concatenate((v[:-1], [v[0]]))
    angles = np.concatenate((angles, [angles[0]]))
    # 绘制折线图
    ax.plot(angles, center, 'o-', linewidth=1, label = "第%d簇人群,%d人"% (i+1, v[-1]))
    # 填充颜色
    ax.fill(angles, center, alpha=0.25)
    # 添加每个特征的标签
    ax.set_thetagrids(angles * 180 / np.pi, feature, fontsize = 9)
    # 设置雷达图的范围
    ax.set_ylim(min_value - 0.1, max_value + 0.1)
    # 添加标题
    plt.title('客户群特征分析图', fontsize=12)
    # 添加网格线
    ax.grid(True)
    # 设置图例
    plt.legend(loc='upper right', bbox_to_anchor=(1.3,1.0), ncol=1, fancybox=True, shadow=True)
    
# 显示图形
plt.show()
```

```python
# 详细输出原始数据及其类别
result = pd.concat([air_data, pd.Series(kmodel.labels_, index=air_data.index)], axis=1)  # 详细输出每个样本对应的类别
result.columns = list(air_data.columns) + ['class']  # 重命名表头
result.to_excel('./result_analysis.xlsx', index=0)  # 保存结果
```
